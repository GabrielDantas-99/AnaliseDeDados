# Funções

```python
# Funcao sem parametro
def funcao():
    print("Olá")
funcao()
```

```python
# Funcao com parametro
def funcao(frase):
    print(frase)
funcao("olá")
```

```python
# Funcao Default
def funcao(inic=1, fim=10):
    for i in range(inic, fim+1):
        print(i)
print("Sem alterar:")
funcao() # Os parametros serao setados como default
print("Alterando os Default:")
funcao(2, 8)
```

# Dicionários (Chave:Valor)

```python
# Declarando um dicionario
notas = {'Joao' : 6.0, 'Maria' : 8.0, 'Pedro' : 6.5}
print(notas)
```

```python
# Sabendo quais as chaves sao acessiveis no dicionario
notas.keys()
```

```python
# Sabendo todos os valores acessiveis no dicionario
notas.values()
```

```python
# Acessando um valor unico no dicionario
print(notas['Joao'])
```

```python
# Verificando a existencia de uma chave
print('Joao' in notas)
print('Fernando' in notas)
```

```python
# Adicionando um novo elemento(chave/valor) no dicionario
notas['Ana'] = 9.0 # Valor é sempre adicionado ao final do dicionario
print(notas)
```

```python
# Apagando um elemento do dicionario
del notas['Joao']
print(notas)
```

```python
# Buscando um valor, caso não encontre mostra segundo parametro
notas.get('Geraldo', "Não Encontrado")
notas.get('Ana', "Não Encontrado")
```

# Numpy

```python
# Importando o Numpy
import numpy as np
```

```python
# Criando uma matriz unidimensional
unid = np.array([12, 24, 26, 18, 10])
print(unid)
print(type(unid))
```

```python
# Criando arrays e setando seus types
arfloat = np.array([1.3, 4.2, 6.4], dtype = np.float64) # Tipo float
print(arfloat); print(type(arfloat))
arint = np.array([1, 3, 5], dtype = np.int32) # Tipo inteiro
print(arint); print(type(arint))
```

```python
# Mudando o tipo do array astype
mtnew = np.array([1.4, 3.6 -5.1, 9.42])
print(mtnew)
newmtint = mtnew.astype(np.int32)
print(newmtint)
oldmtint = newmtint.astype(np.float64)
print(oldmtint)
```

```python
# Estrutura Bidimensional
mt7 = np.array([[7,2,4], [6,4,1], [9,6,3]])
print(mt7)
# Acessando elemento especifico
print(mt7[0][1])
# Tamanho das dimensões da matriz
print(mt7.shape)
```

```python
# Funções matematicas
print(mt7.max()) # Valor máximo
print(mt7.min()) # Valor minimo 
print(mt7.sum()) # Soma os valores
print(mt7.mean()) # Media dos valores
print(mt7.std()) # Desvio padrao 
```

```python
# Criando arrays vazios tipificados
vazio = np.empty([3,2], dtype=int) # Empty: não são inicializados, mas nao sao vazios
vazio
```

```python
# Criando uma matriz 4x3 com valores 0
zeros = np.zeros([4,3]) # Com valores igual a um: np.ones([])
zeros
```

```python
# Criando uma matriz com diagonal principal com valores 1
diagonal = np.eye(5)
diagonal
```

```python
# Valores aleatorios entre 0 e 1
n2 = np.random.random((5))
print(n2)
# Numero aleatorios com distruição normal contendo negativos
n3 = np.random.randn((5))
print(n3)
```

```python
# Multiplicando array aleatorio 3 x 4
ar2 = 10 * np.random.random((3,4))
print(ar2)
```

```python
# Gerando numeros aleatorios com uso de semente(Mesmos nums ale nas execuções)
gnr = np.random.default_rng(1) # Ao mudar esse numero serao gerados outrs aleatorios
ale5 = gnr.random(3)
print(ale5)

# Gerando inteiros
ale6 = gnr.integers(10, size = (3, 4))
print(ale6)
```

```python
# Removendo repetições 
j = np.array([11, 12, 13, 14, 15, 16, 13, 11, 20])
j = np.unique(j)
print(j)
```

```python
# Adição e multiplicação de matrizes
n = np.array([[1, 2], [3, 4]])
m = np.array([[5, 6], [7, 8]])
soma = n + m
print(soma)
print('-------')
multiplicacao = n * m
print(multiplicacao)
print('-------')
o = np.array([2, 4])
mult = m * o
print(mult)
```

```python
# Transposição com o método T
f = np.arange(15).reshape(3, 5)
print(f)
print('-------')
s = f.T
print(s)
```

```python
# Expressões lógicas
v = np.random.randn(4, 4)
print(v)
# Criando matriz de valores bool baseada no array v
x = v > 0
print(x)
# Criando matriz com 1 e -1 baseada nos valores do array x
z = np.where(x > 0, 1, -1)
print(z)
```
